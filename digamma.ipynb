```python
import sympy
from sympy import log, exp, integrate, pi, symbols, simplify
from sympy.functions.special.gamma_functions import digamma, gamma
from sympy.plotting import plot
```

```python
x = symbols('x')
f = digamma(-1/x)
```

```python
plot(f, (x, 1.77, 5))
```

```python
plot(f, (x, -10, -1e-5))
```

```python
plot(f, (x, -100, -10))
```

```python
[(i, v, v / i) for i in xrange(2,25) for v in [f.evalf(subs={x: i})]]
```

```python
f.evalf(subs={x:1e+15}) / 1e+15
```

```python
plot(f, (x, -3, -0.5))
```
