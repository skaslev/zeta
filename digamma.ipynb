```python
import sympy
from sympy import symbols
from sympy.functions.special.gamma_functions import digamma
from sympy.plotting import plot

x = symbols('x')
f = digamma(-1/x)
```

```python
plot(f, (x, 1.77, 5))
```

```python
plot(f, (x, -10, -1e-5))
```

```python
plot(f, (x, -100, -10))
```

```python
[(i, f.evalf(subs={x: i})) for i in xrange(2,15)]
```

```python
f.evalf(subs={x:1e+15}) / 1e+15
```
