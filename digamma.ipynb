```python
def newton(f, x, x0, eps=1e-15):
    c = x0
    while True:
        if abs(f.evalf(subs={x: c})) < eps:
            break
        c += -(f / f.diff(x)).evalf(subs={x: c})
    return c
```

```python
[[r, fr] for r in [newton(f, x, 2)] for fr in [f.evalf(subs={x: r})]]
```
